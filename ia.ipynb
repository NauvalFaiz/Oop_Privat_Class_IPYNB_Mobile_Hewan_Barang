```python
# Import library
import matplotlib.pyplot as plt

# Data contoh (Top 10 populasi dunia)
countries = [
    "China", "India", "USA", "Indonesia", "Pakistan",
    "Brazil", "Nigeria", "Bangladesh", "Russia", "Mexico"
]

population = [
    1412, 1408, 331, 277, 240,
    216, 223, 173, 146, 128
]  # dalam juta

# Warna (Top 3 beda warna)
colors = ["red" if i < 3 else "skyblue" for i in range(len(countries))]

# Plot
plt.figure(figsize=(10,6))
plt.barh(countries, population, color=colors)

# Label
plt.xlabel("Population (Millions)")
plt.title("Top 10 Most Populated Countries")

# Balik urutan biar terbesar di atas
plt.gca().invert_yaxis()

# Tampilkan angka di bar
for i, v in enumerate(population):
    plt.text(v + 10, i, str(v), va='center')

plt.show()
```
