### Latihan Kucing
---

```python
class Kucing:
    def __init__(self, nama, warna):
        self.nama = nama
        self.warna = warna
    def suara(self):
        return f"{self.nama} Berkata: aku Prabowi"
Kucing1 = Kucing("Jokwi", "Merah")
Kucing2 = Kucing("Megi","Red")

print(Kucing1.nama)
print(Kucing2.suara())
        
```
